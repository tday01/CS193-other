**Image Alterations & Manipulations**

```python
# envo dependencies: matplotlib, pillow, tensorflow, scikit-image, numpy
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from matplotlib.pyplot import imread, imshow, subplots, show
import tensorflow as tf

def plot(data_generator):
    data_generator.fit(images)
    image_iterator = data_generator.flow(images)
    
    # Plot the images given by the iterator
    fig, rows = subplots(nrows=1, ncols=4, figsize=(18,18))
    for row in rows:
        row.imshow(image_iterator.next()[0].astype('int'))
        row.axis('off')
    show()
    
image = imread("image.jpg")

# Creating a dataset which contains just one image.
images = image.reshape((1, image.shape[0], image.shape[1], image.shape[2]))

imshow(images[0])
print('Original')
show()
```

**Image Manipulations**

```python
# rotation

data_generator = ImageDataGenerator(rotation_range=90)
plot(data_generator)
```

```python
# brightness

data_generator = ImageDataGenerator(brightness_range=(0.1, 0.9))
plot(data_generator)
```

```python
# shear intensity

data_generator = ImageDataGenerator(shear_range=45.0)
plot(data_generator)
```

```python
# channel shift

data_generator = ImageDataGenerator(channel_shift_range=150.0)
plot(data_generator)
```

```python
# horizontal flip

data_generator = ImageDataGenerator(horizontal_flip=True)
plot(data_generator)
```

```python
# vertical flip

data_generator = ImageDataGenerator(vertical_flip=True)
plot(data_generator)
```

**Image Effects**

```python
from skimage import data, img_as_float
from skimage import exposure
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# Histogram Equalization

img_eq = exposure.equalize_hist(image)
imgplot = plt.imshow(img_eq)
```

```python
# Contrast stretching

p2, p98 = np.percentile(image, (2, 98))
img_rescale = exposure.rescale_intensity(image, in_range=(p2, p98))

imgplot = plt.imshow(img_rescale)
```

```python
# Adaptive Equalization

img_adapteq = exposure.equalize_adapthist(image, clip_limit=0.03)
imgplot = plt.imshow(img_adapteq)
```

```python
# spectral

imgplot = plt.imshow(image)
imgplot.set_cmap('nipy_spectral')
```

```python
# Interpolation

imgplot = plt.imshow(image, interpolation="nearest")
```

**Edge Enhancement**

```python
from PIL import Image
from PIL import ImageFilter

image = Image.open("./image.jpg")
```

```python
# Apply edge enhancement filter

edgeEnahnced = image.filter(ImageFilter.EDGE_ENHANCE)
imgplot = plt.imshow(edgeEnahnced)
```

```python
# Apply increased edge enhancement filter

moreEdgeEnahnced = image.filter(ImageFilter.EDGE_ENHANCE_MORE)
imgplot = plt.imshow(moreEdgeEnahnced)
```
